```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
page=requests.get('https://www.flipkart.com/search?q=smartphone&as=on&as-show=on&otracker=AS_Query_OrganicAutoSuggest_1_5_na_na_na&otracker1=AS_Query_OrganicAutoSuggest_1_5_na_na_na&as-pos=1&as-type=HISTORY&suggestionId=smartphone&requestId=db9970d4-ebd2-4d32-829d-d4a533eb803d')
page
```

```python
soup=BeautifulSoup(page.content)
soup
```

```python
BrandName=[]

for i in soup.find_all('div', class_="_4rR01T"):
    BrandName.append(i.text)
    
BrandName
  

```

```python
RAMandROM=[]
for i in soup.find_all('div',class_="fMghEO" ):
    RAMandROM.append(i.text)
RAMandROM
```

```python
Price=[]
for i in soup.find_all('div', class_="_30jeq3 _1_WHN1"):
    Price.append(i.text)
Price
```

```python
URL=[]
for i in soup.find_all('a',class_="_1fQZEK"):
    URL.append(i['href'])
URL
```

```python
print(len(BrandName),len(RAMandROM),len(Price),len(URL))
```

```python
df=pd.DataFrame({'Brand and Smartphones':BrandName,'Specifations':RAMandROM,'Product price':Price,'Product URL':URL})
df
```

```python
df.to_csv('Smartphones.csv', index=False)
```

9.hostelworld

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
page=requests.get('https://www.hostelworld.com/st/hostels/europe/england/london')
page
```

```python
soup=BeautifulSoup(page.content)
soup
```

```python
Name=[]
for i in soup.find_all('div', class_="property-name"):
    Name.append(i.text)
Name
```

```python
ratings=[]
for i in soup.find_all('div',class_="rating-score"):
    ratings.append(i.text)
ratings
```

```python
reviews=[]
for i in soup.find_all('div', class_="review"):
    reviews.append(i.text)
reviews
```

```python
distance=[]
for i in soup.find_all('div',class_="property-distance"):
    distance.append(i.text)
distance
```

```python
description=[]
for i in soup.find_all('div',class_="property-description"):
    description.append(i.text)
description
```

```python
dorms=[]
for i in soup.find_all('strong',class_="current"):
    dorms.append(i.text)
dorms
```

```python
print(len(Name),len(ratings),len(dorms),len(distance),len(reviews),len(description))
```

6. digit.in

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
page=requests.get('https://www.digit.in/?s=laptops&post_type=product')
page
```

```python
soup=BeautifulSoup(page.content)
soup
```

```python
brandname=[]
for i in soup.find_all('h3', class_="text-clamp text-clamp-2"):
    brandname.append(i.text)
brandname
```

```python
ratings=[]
for i in soup.find_all('div',class_="rh_woo_star"):
    ratings.append(i.text)
ratings
```

```python
amount=[]
for i in soup.find_all('span',class_="woocommerce-Price-amount amount"):
    amount.append(i.text)
amount
```

```python
print(len(brandname),len(ratings),len(amount))
```
