2.Former Presidents of India

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
page=requests.get('https://presidentofindia.nic.in/former-presidents')
```

```python
page
```

```python
soup= BeautifulSoup(page.content)
soup
```

```python
name=[]

for i in soup.find_all('div',class_="desc-sec"):
    name.append(i.text)
    
name
```

```python
import pandas as pd
```

```python
presidents_df = pd.DataFrame([
    {'Name': info.split('\n')[1].strip(), 'Term of Office': info.split('\n')[2].strip()}
    for info in name
])


print("List of Respected Former Presidents of India:")
print(presidents_df)
```

7.Dine out

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
page= requests.get('https://www.dineout.co.in/delhi-restaurants/buffet-special')
```

```python
page
```

```python
soup=BeautifulSoup(page.content)
soup
```

```python
RestaurantNames=[]

for i in soup.find_all('div', class_="restnt-info cursor"):
    RestaurantNames.append(i.text)
    
RestaurantNames    
```

```python
locations=[]

for i in soup.find_all('div', class_="restnt-loc ellipsis"):
    locations.append(i.text)

locations    
```

```python
cuisine=[]

for i in soup.find_all('span', class_="double-line-ellipsis"):
    cuisine.append(i.text)
cuisine    
```

```python
ratings=[]

for i in soup.find_all('div', class_="restnt-rating rating-4"):
    ratings.append(i.text)
ratings    
```

```python
images=[]
for i in soup.find_all('img', class_="no-img"):
    images.append(i['data-src'])
images    
```

```python
print(len(RestaurantNames),len(locations),len(cuisine),len(ratings),len(images))
```

```python
df=pd.DataFrame({'RestaurantName':RestaurantNames,'location':locations,'cuisine and price':cuisine,'ratings':ratings,'images_url':images})
df
```
